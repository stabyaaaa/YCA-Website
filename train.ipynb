```python
import time

# Epoch durations in seconds
epoch_times = [
    10,        # Epoch 1 - 10 sec
    30,        # Epoch 2 - 30 sec
    50,        # Epoch 3 - 50 sec
    600,       # Epoch 4 - 10 min
    3600,      # Epoch 5 - 1 hr
    7200,      # Epoch 6 - 2 hr
    10800,     # Epoch 7 - 3 hr
    14400,     # Epoch 8 - 4 hr
    21600,     # Epoch 9 - 6 hr
    36000      # Epoch 10 - 10 hr
]

def load_data():
    print("Loading dataset...")
    time.sleep(2)

def preprocess():
    print("Preprocessing data...")
    time.sleep(2)

def train_step():
    print("Forward pass")
    time.sleep(1)

    print("Compute loss")
    time.sleep(1)

    print("Backward pass")
    time.sleep(1)

    print("Optimizer step")
    time.sleep(1)

def validate():
    print("Running validation...")
    time.sleep(3)

# Start pipeline
load_data()
preprocess()

for epoch in range(10):

    print(f"\n==========================")
    print(f"Epoch {epoch+1}/10 started")
    print(f"Target time: {epoch_times[epoch]} seconds")
    print(f"==========================")

    start_time = time.time()

    # Simulated training batches
    for step in range(5):
        print(f"Batch {step+1}/5")
        train_step()

    validate()

    # Ensure epoch lasts exact required duration
    elapsed = time.time() - start_time
    remaining = epoch_times[epoch] - elapsed

    if remaining > 0:
        print(f"Waiting {int(remaining)} sec to match epoch duration...")
        time.sleep(remaining)

    total = time.time() - start_time
    print(f"Epoch {epoch+1} finished in {int(total)} seconds")

print("\nTraining completed!")
```
